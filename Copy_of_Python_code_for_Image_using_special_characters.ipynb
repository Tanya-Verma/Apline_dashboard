````python
import json

def decode(gray_file='/content/drive/MyDrive/55a87950 (1).txt', map_file='/content/drive/MyDrive/33391b5d.json', output='/content/drive/MyDrive/75209d59.txt'):
    with open(map_file) as f:
        mapping = json.load(f)['char_to_gray']
    gray_to_char = {v: k for k, v in mapping.items()}

    decoded_lines = []
    with open(gray_file) as f:
        for line in f:
            nums = [int(x) for x in line.split()]
            chars = [gray_to_char.get(n, '?') for n in nums]
            decoded_lines.append(''.join(chars))

    with open(output, 'w') as f:
        f.write('\n'.join(decoded_lines))
    print(f"Decoded art saved to {output}")

# Run: decode()
````

````python
import json
from IPython.display import display, Markdown

def decode_and_print_ascii(gray_file='/content/drive/MyDrive/55a87950.txt', map_file=None, output='/content/drive/MyDrive/75209d59.txt'):
    """
    Decode grayscale matrix to ASCII art and print in Google Colab.
    Uses conversation mapping or standard greyscale if no map_file.
    """
    # Standard greyscale mapping (dark to light) if no mapping.json
    if map_file is None:
        standard_gray_to_char = {
            0: '@', 32: '%', 77: '#', 106: '*', 121: '+', 149: '=', 163: '-', 231: '.'
        }
        gray_to_char = standard_gray_to_char
        print("Using standard greyscale mapping (no mapping.json)")
    else:
        with open(map_file) as f:
            mapping = json.load(f)['char_to_gray']
        gray_to_char = {v: k for k, v in mapping.items()}
        print("Using mapping.json")

    # Decode from your attached 55a87950.txt
    decoded_lines = []
    with open(gray_file, 'r') as f:
        for line in f:
            line = line.strip()
            if not line: continue
            nums = [int(x) for x in line.split()]
            chars = [gray_to_char.get(n, '?') for n in nums]
            decoded_lines.append(''.join(chars))

    # Save full
    with open(output, 'w') as f:
        f.write('\n'.join(decoded_lines))

    # Print in Colab: use Markdown for monospace fixed-width
    art_md = '```\n' + '\n'.join(decoded_lines) + '\n```'
    display(Markdown(art_md))

    print(f"âœ… Full ASCII art saved to {output}")
    return decoded_lines

# RUN THIS IN GOOGLE COLAB:
decode_and_print_ascii(gray_file='/content/drive/MyDrive/55a87950 (1).txt')  # Uses your attached file!
````

````python
import json
from IPython.display import display, Markdown

def decode_and_print_ascii_loops(gray_file='/content/drive/MyDrive/55a87950 (2).txt', map_file=None, output='/content/drive/MyDrive/75209d59.txt'):
    """
    Decode grayscale matrix to ASCII art using FOR LOOPS for printing.
    Perfect for Google Colab - loops over rows/columns explicitly.
    """
    # Standard greyscale mapping (dark to light) if no mapping.json
    if map_file is None:
        standard_gray_to_char = {
            0: '@', 32: '%', 77: '#', 106: '*', 121: '+', 149: '=', 163: '-', 231: '.'
        }
        gray_to_char = standard_gray_to_char
        print("âœ… Using standard greyscale mapping (loops handle all chars)")
    else:
        with open(map_file) as f:
            mapping = json.load(f)['char_to_gray']
        gray_to_char = {v: k for k, v in mapping.items()}
        print("âœ… Using mapping.json")

    # Decode using loops: read file â†’ loop lines â†’ loop numbers â†’ map to chars
    decoded_lines = []
    line_index = 0
    with open(gray_file, 'r') as f: # Corrected: Ensure this path is absolute and correct
        for line in f:  # LOOP 1: Over file lines (rows)
            line = line.strip()
            if not line:
                continue

            # LOOP 2: Split numbers in row
            num_strs = line.split()
            row_chars = []
            for num_str in num_strs:  # LOOP 3: Over numbers in row
                num = int(num_str)
                ch = gray_to_char.get(num, '?')  # Map grayscale â†’ char
                row_chars.append(ch)  # Collect chars

            decoded_line = ''.join(row_chars)
            decoded_lines.append(decoded_line)
            line_index += 1

    # Save using loop
    with open(output, 'w') as f:
        for i, line in enumerate(decoded_lines):  # LOOP 4: Write lines
            f.write(line + '\n')

    print(f"âœ… Decoded {len(decoded_lines)} lines â†’ saved to {output}")

    # PRINT using NESTED LOOPS in Colab (monospace via Markdown)
    art_lines = []
    for row_idx in range(len(decoded_lines)):  # LOOP 5: Rows
        line = decoded_lines[row_idx]
        row_display = []
        for col_idx in range(len(line)):  # LOOP 6: Columns/chars
            row_display.append(line[col_idx])  # Access each char
        art_lines.append(''.join(row_display))

    # Colab display
    art_md = '```\n' + '\n'.join(art_lines) + '\n```'
    display(Markdown(art_md))

    return decoded_lines

# ðŸš€ RUN THIS IN GOOGLE COLAB (upload 55a87950.txt first):
decode_and_print_ascii_loops(gray_file='/content/drive/MyDrive/55a87950.txt') # Corrected: Use the absolute path to the file in Google Drive
````
